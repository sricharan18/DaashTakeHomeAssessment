```python
import pandas as pd
```

```python
df = pd.read_csv('Lowes_data.csv', na_values=['NA'])
```

```python
df
```

# No.of Wreaths per Brand

```python
pd.DataFrame(df['Brand'].value_counts())
```

# Price range for wreaths

```python
df.groupby("Brand").agg({'Price': ['max', 'min']}).reset_index()
```

# Histogram of product prices

```python
import matplotlib.pyplot as plt

plt.hist(df['Price'], bins=20) 
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.title('Histogram of Product Prices')
plt.show()
```

# Most frequently occurring features and keywords from the product descriptions

```python
from wordcloud import WordCloud

all_descriptions = " ".join(df['Description'].dropna())

wordcloud = WordCloud(width=800, height=400, background_color='white').generate(all_descriptions)

plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Feature Popularity Analysis - Word Cloud', fontsize=16)
plt.show()
```
